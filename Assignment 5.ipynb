#### 2. Replace all None or NaN with 0

#### 3. Drop the column if more than 90% value in this column is 0. (Hint, using pd.apply() function)

#### 4. Normalize the table (columns from Accumulated Other Comprehensive Income (Loss) to Selling, General and Administrative Expenses)

Using pd.apply() to normalize the table, in this table, you need to calculate the largest value, and smallest value in each column, compute the range for each variable. Then using value in this column divided by range of this column. (Do not using any function like MinMax(), you need to write it by yourself)

```python
def normalize(column):
    # Find the max and min value
    max_x = column[0]
    min_x = column[0]
    for item in column:
        if item < min_x:
            min_x = item
        if item > max_x:
            max_x = item
    
    # Calculate the normalized value and return transformed values
    for i in range(len(column)):
        column[i] = column[i]/(max_x - min_x)
    
    return column
```

```python
BalanceSheet.loc[:,'Accumulated Other Comprehensive Income (Loss)':'Selling, General and Administrative Expenses'] = BalanceSheet.loc[:,'Accumulated Other Comprehensive Income (Loss)':'Selling, General and Administrative Expenses'].apply(normalize)
```

#### 6. If you look at column ('Company Name'), you will find some company name end with 'CORP', 'CO' or 'INC'. Create a new column (Name: 'CO') to store the last word of company name. (For example: 'CORP', 'CO' or 'INC') (Hint: using map function)

```python
BalanceSheet['CO'] = BalanceSheet['Company Name'].str.split(" ").apply(lambda x: x[-1])
```

#### 7. Merge (inner) Ratings and BalanceSheet based on 'datadate' and 'Global Company Key', and name merged dataset 'Matched'.

```python
Matched = pd.merge(BalanceSheet, Ratings, on = ['Data Date', 'Global Company Key'], how = 'inner')
```
