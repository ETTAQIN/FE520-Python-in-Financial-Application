### 1.1 Question1

```python
class Rectangular:
    # assgin length and width in constructor
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    # define area function
    def area(self):
        area = self.length * self.width
        return area
            
    # define perimeter function
    def perimeter(self):
        perimeter = 2*(self.length + self.width)
        return perimeter
```

```python
# set length as 10, and width as 20
myRec = Rectangular(10, 20)
```

```python
print(myRec.area())
```

```python
print(myRec.perimeter())
```

#### Inherited class

```python
class Square(Rectangular):
    def __init__(self, long):
        Rectangular.__init__(self, long, long)
```

```python
mySqu = Square(10)
```

```python
print(mySqu.area())
```

```python
print(mySqu.perimeter())
```

```python
class Time:
    
    def __init__(self, hours, minutes, seconds):
        self.hours = hours
        self.minutes = minutes
        self.seconds = seconds
    
    def addTime(self, hours, minutes, seconds):
        # Directly add time by hour, minute and second separately
        hr = self.hours + hours
        minute = self.minutes + minutes
        second = self.seconds + seconds
        
        # calculate mod of seconds to get new seconds
        sec = second % 60
        
        # Add minutes and hours generated by seconds and minutes seperately
        if (second // 60) > 0:
            minute += (second // 60)
            mins = minute % 60
            if (minute // 60) > 0:
                hr += (minute // 60)
        
        return hr, mins, sec
       
    def displayTime(self):
        print('{hour}hours {minutes}minutes {seconds}seconds'.format(hour = self.hours, minutes = self.minutes, seconds = self.seconds))
    
    def DisplayMinute(self):
        seconds = self.hours *3600 + self.minutes * 60 +self.seconds
        return seconds
```

## 3. Find stationary point

#### 3.1 Find the solution of this stationary for initial conditions $x_0 = 0$, $x_0 = 4$.
$$ f'(x) = x +4 - exp(x)$$

```python
# Define the function f
f = lambda x: x + 4 - np.exp(x)

# Basedon Euler method to find out the stationary points
def euler(x0, ita, f):
    # Calculate the initiate x1
    x1 = x0 - ita*f(x0)
    
    while np.abs(x0-x1)>10**(-6):
        x0 = x1
        x1 = x0 - ita*f(x0)
        if x0 == np.inf:
            x1 = x0
            break
    return x1
```

When $x_0 = 0$

```python
euler(0, 0.01, f)
```

When $x_0 = 4$

```python
euler(4, 0.01, f)
```

#### 3.2 Find the stationary point of the function (Bisection method)
$$ f'(x)=sin(x)exp(x)$$

```python
def bisection(a, b):
    # Define the function
    def f(x):
        return np.sin(x)*np.exp(x)
    
    while np.abs(b-a)>10**(-6):
        # find the middle point
        m = (a+b)/2
        if f(a)*f(m) < 0:
            b = m
        else:
            a = m
    return b    
```

```python
bisection(2,4)
```

#### 3.3 Using a numerical solvor find the fixed points, both stable and unstable for the following differential equation, with $x\in [-4;4]$:
$$f'(x) = x\sin(x) -1$$

```python
# Divide x into small intervals with 1
x = np.arange(-4,4,1)
x
```

```python
def g(x):
    return x*np.sin(x)-1

def euler2(x0, ita):
    x1 = x0 - ita*g(x0)
    while np.abs(x0-x1)>10**(-3):
        x0 = x1
        x1 = x0 - ita*g(x0)
        if x0 == np.inf:
            x1 = x0
            break
    return x1
```

```python
stable_points = []
for i in x:
    stable = euler2(i, 0.01)
    stable_points.append(stable)
```

```python
new_stable_points = []

for i in range(len(stable_points)-1):
    if np.abs(stable_points[i]-stable_points[i+1]) > 10**(-1):
        new_stable_points.append(stable_points[i])
        i += 1   
    else:
        i += 1
new_stable_points = new_stable_points[0:2]
new_stable_points
```

```python
unstable_points = []
for i in x:
    unstable = euler2(i, -0.01)
    unstable_points.append(unstable)
unstable_points
```

```python
new_unstable_points = []

for i in range(len(unstable_points)-1):
    if np.abs(unstable_points[i]-unstable_points[i+1]) > 10**(-1):
        new_unstable_points.append(unstable_points[i+1])
        i += 1   
    else:
        i += 1
new_unstable_points = new_unstable_points[1:3]
new_unstable_points
```
